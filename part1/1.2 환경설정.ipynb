## init_chat_model

```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# .env 파일에서 환경 변수 로드
load_dotenv()

# init_chat_model 사용 (권장)
gpt4_mini = init_chat_model(
    "openai:gpt-4.1-mini",
    temperature=0.7,
    max_tokens=150,
)

gpt4 = init_chat_model(
    "openai:gpt-4.1",
    temperature=0.7,
    max_tokens=300,
)
```

## 모델 사용예제

```python
from langchain_core.messages import HumanMessage

# GPT-4.1-mini 사용
response_mini = gpt4_mini.invoke([HumanMessage(content="Hello, how are you?")])
print(response_mini.content)

# GPT-4.1 사용
response_full = gpt4.invoke([HumanMessage(content="Explain the concept of machine learning.")])
print(response_full.content)
```

## 멀티 프로바이더 설정

```python
import sys
print(sys.executable)


#1. 모델 준비
llama3 = init_chat_model("ollama:llama3.2")

# 2. 사용자 질문
query = "What is the weather in SF?"

import pprint
# 3. LLM 호출 
msg = llama3.invoke(query)
pprint.pprint(msg.content)
```

## rich.pretty

```python
pprint.pp(msg)
pprint.pprint(msg)


from rich.pretty import pprint as rpp
rpp(msg)
rpp(msg.content)
```
